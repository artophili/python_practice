```python
def merge_dicts_with_overlapping_keys(dicts):
    merged_dict = {}
    for i in dicts:
        for j in i.keys():
            if j in merged_dict:
                merged_dict[j] = i[j] + merged_dict[j]
            else:
                merged_dict[j] = i[j]
    return merged_dict

m = merge_dicts_with_overlapping_keys([{'a': 1, 'b': 2}, {'b': 3, 'c': 4}, {'c': 5, 'd': 6}])
print(m)
```
